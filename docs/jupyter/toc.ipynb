## [HTM architecture](theory.ipynb)

Short description of the modules and ideas behind HTM theory.

#### [Encoders](encoders.ipynb)

Converts environment data into SDP, which is used internally in the system.

#### [Spatial Pooler (Mapper)](spatial_mapper.ipynb)

Normalizes the output from the Encoders.

#### [Temporal Memory design](tm_design.ipynb)

Prerequisite for Segmented Memory and  Temporal Memory

#### [Segmented Memory (Classifier)](segmented_memory.ipynb)

Memorizes SDP =to=> SDP patterns.
Can be used as Classifier and is the base on top of which TM is implemented.

#### [Temporal Memory](temporal_memory.ipynb)

Learns and Predicts variable-order sequence by remembering higher dimensional transitions.

#### LexTM

Temporal memory with some lexicon related functionality.

#### Temporal Pooler

Transform temporal sequence to stable and unique spatial pattern that is used to communicate with other modules.
